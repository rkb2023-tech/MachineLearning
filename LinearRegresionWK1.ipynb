```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Download and prepare the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
```

```python
lifesat.head()
```

```python
X.shape
```

```python
y.shape
```

```python
# Visualize the data
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
```

```python
# Select a linear model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
X_new = [[x] for x in range(int(X.min()), int(X.max()) + 1000, 1000)]
y_pred = model.predict(X_new)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Linear regression line')
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life satisfaction")
plt.title("Linear Regression: GDP vs Life Satisfaction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Make a prediction for Cyprus
X_new = [[37_655.2]]  # Cyprus' GDP per capita in 2020
print(model.predict(X_new)) # output: [[6.30165767]]
```

#Knearest Neighbour

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=3)
```

```python
model.fit(X, y)
```

```python
# Predict values
X_new = [[x] for x in range(int(X.min()), int(X.max()) + 1000, 1000)]
y_pred = model.predict(X_new)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_new, y_pred, color='green', linewidth=2, label='KNN regression (k=3)')
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life satisfaction")
plt.title("KNN Regression: GDP vs Life Satisfaction")
plt.legend()
plt.grid(True)
plt.show()
```
